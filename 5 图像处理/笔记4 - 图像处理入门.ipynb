## 图片是如何表达的？

计算机是离散的，因此存储的图片也是离散的，图片离散的基本单元就是像素点（pixel），一般像素点的颜色都是由红绿蓝三色通道构成的，也就是常说的RGB，每个通道的强度用一个uint8表示（也就是一个0-255的整数，见笔记1）。不同的RGB组合会产生不同的颜色。这样一张图片就可以被表达成长$\times$宽$\times$3的一个数组。而我们对该数组进行操作就可以对图像进行操作了。

```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt

def disp(a, title):
    plt.subplot(a)                        # create subplot
    plt.xticks([],[]); plt.yticks([],[])  # remove ticks
    plt.title(title)

img = Image.open('lenna.png')
plt.figure(figsize=(16,4))
I = np.array(img)
print("Shape of Image: \t"+str(I.shape))
print("Data Type in Array:\t"+str(I.dtype))

disp(141, "Original")
plt.imshow(I)

disp(142,"Red Channel")
plt.imshow(I[:,:,0], cmap=plt.cm.Reds_r)

disp(143,"Green Channel")
plt.imshow(I[:,:,1], cmap=plt.cm.Greens_r)

disp(144,"Blue Channel")
plt.imshow(I[:,:,2], cmap=plt.cm.Blues_r)
plt.show()
```

## 灰度

灰度是像素点的明暗程度。由于景物各点的颜色及亮度不同，摄成的黑白照片上或电视接收机重现的黑白图像上各点呈现不同程度的灰色。把白色与黑色之间按对数关系分成若干级，称为“灰度等级”。范围一般从0到255，白色为255，黑色为0，故黑白图片也称灰度图像。

灰度没有一个确定的计算方法，但普遍来说以下两种计算方法会比较常用

- $gray=(R+G+B)\div3$

- $gray=(R\times299+G\times587+B\times114+500)\div1000$

（*注意：由于RGB都是uint8，计算以上式子可能会遇到整数溢出的问题（见笔记1整数部分），因此为了避免该问题，我们应该将其扩充成更大的整数比如uint32、int64，以防止溢出*）

```python
# overflow
R,G,B = I[:,:,0],I[:,:,1],I[:,:,2]
Gray0 = (R+G+B)//3
Gray1 = (R*299+G*587+B*114+500)//1000

plt.figure(figsize=(8,4))
disp(121, r"$gray=\frac{R+G+B}{3}$")
plt.imshow(Gray0, cmap=plt.cm.gray)

disp(122, r"$gray=\frac{299R+587G+114B+500}{1000}$")
plt.imshow(Gray1, cmap=plt.cm.gray)
plt.show()
```

```python
# overflow-proof
I64 = np.int_(I) #convert unit8 to int64
R,G,B = I64[:,:,0],I64[:,:,1],I64[:,:,2]
Gray0 = (R+G+B)//3
Gray1 = (R*299+G*587+B*114+500)//1000

plt.figure(figsize=(8,4))
disp(121, r"$gray=\frac{R+G+B}{3}$")
plt.imshow(Gray0, cmap=plt.cm.gray)

disp(122, r"$gray=\frac{299R+587G+114B+500}{1000}$")
plt.imshow(Gray1, cmap=plt.cm.gray)
plt.show()
```

## 隐写术

事实上，我们读取的这张`Lenna.png`内有玄机，这个将作为我们的作业。

回顾一下，一个像素点是由RGB三种颜色组成的，每个通道的强度均由uint8表示，假定某个通道的强度为112，其二进制表达为`0111 0000`，那么根据连续性原理，这个颜色应当与强度为113的颜色非常接近，也就是说如果我们将该通道的末位做一些更改，对肉眼来说，图片并不会有什么明显改变。于是我们就可以考虑利用这个性质将信息写入图片。

比如，我们可以将一张图片（有损）写入另一张图片：因为对于uint8来说，前4位对强度的影响要远高于后4位，于是我们可以将被写入的图片的每个像素的前4位写到被写入的图片的后4位上，形成一张“带有隐藏信息”的图片，当我们需要读取图片中信息时，就可以取出每个像素点后4位放到前4位上，即可恢复隐藏信息。这种技术被称为“隐写术”。

比如，用于隐藏信息的图片的某个像素点，其强度为112，也就是`0111 0000`，而需要被隐藏的图片对应的像素点强度为231，也就是`1110 0111`，我们取出231的前4位`1110`，放在112的后4位上，形成新的强度`0111 1110`(126)，而126和112相差仅14，颜色上近似。因此经过以上过程的处理，得到的图片和原图应该很接近，肉眼上的差别并不会特别大。比如以下案例中，我们尝试将一张宇宙的图片写入Lenna：

```python
I = np.array(Image.open('lenna.png'))
I1 = np.array(Image.open('galaxy.jpg'))

newI = ((I//2**4)*(2**4) + (I1//2**4))

recoverI = newI % (2**4) * (2**4)

plt.figure(figsize=(12,4))
disp(131,"Original")
plt.imshow(I)

disp(132,"With Hidden Watermark")
plt.imshow(newI)

disp(133,"Hidden Image")
plt.imshow(recoverI)
plt.show()
```

可以看到，处理后的图像和原图除了细节略有差别外，整体观感是没有区别的，如果我们处理后的图片的从后4位提取信息的话，就可以得到图片中“隐藏的宇宙”。这个技术通常被应用于版权保护，将水印图片写入被保护的图片中，比如网上商城的卖家上传照片时，商城就会对图片进行处理，打上该商家的水印，如果有人盗用他的图片，通过隐写写入的水印图片，就可以知道谁是图片的著作人了。类似的，视频和音频实际上从本质上说也是数，也可以利用相似的思路进行隐写，实现产权保护。
