```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt

img = Image.open('1.png')
I = np.array(img)
img = Image.open('galaxy.jpg')
I1 = np.array(img)

newI = ((I//2**4)*(2**4) + (I1//2**4))+np.random.randint(1,15,size=(256,256,3),dtype=np.uint8)
newI = ((I//2**4)*(2**4) + (I1//2**4))//2

recoverI = newI % (2**4) * (2**4)

plt.figure(figsize=(18,6))
plt.subplot(131)
plt.imshow(I)
plt.subplot(132)
plt.imshow(newI)
plt.subplot(133)
plt.imshow(recoverI)
plt.show()
```
